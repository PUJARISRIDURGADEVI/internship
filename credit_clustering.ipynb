```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X= scaler.fit_transform(df)
```

```python
X
```

```python
plt.scatter(X[:, 0], X[:, 1], color='orange') 
plt.xlabel('Alcohol (scaled)')
plt.ylabel('Malic Acid (scaled)')
plt.title('Scaled Alcohol vs Malic Acid')
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k,init="k-means++")
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.xlabel("Number of Clustrers")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
y_pred = kmeans.fit_predict(X)
```

```python
y_pred
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='Set1', s=80)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='black', s=200, marker='X', label='Centers')
plt.xlabel('Alcohol (scaled)')
plt.ylabel('Malic Acid (scaled)')
plt.title('KMeans Clustering: Alcohol vs Malic Acid')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_coefficients=[]
for k in range(2,11):
    kmeans=KMeans(n_clusters=k,init="k-means++")
    kmeans.fit(X)
    score=silhouette_score(X,kmeans.labels_)
    silhouette_coefficients.append(score)
```

```python
silhouette_coefficients
```

```python
plt.plot(range(2,11),silhouette_coefficients)
```
