```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=True, facecolor='g', alpha=0.75)


plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
plt.plot([1, 2, 3, 4])
plt.ylabel('some numbers')
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis((0, 6, 0, 20))
plt.show()
```

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(10, 4))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

```python
y=[1, 2, 3, 4, 5]
x=[1, 2, 3, 4, 5]
plt.plot(x, y, marker='o', color='red', linewidth=2, linestyle='--')
plt.title('Line Plot with Customizations')
```

```python
import seaborn as sns
dots = sns.load_dataset("dots")
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate", col="align",
    hue="choice", size="coherence", style="choice",
    facet_kws=dict(sharex=False),
)
```

```python
fmri = sns.load_dataset("fmri")
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", col="region",
    hue="event", style="event",
    
)
```

```python
tips = sns.load_dataset("tips")
sns.displot(data=tips, x="total_bill", col="time", kde=True,hue="smoker",)
```

```python
tips = sns.load_dataset("tips")
sns.displot(data=tips, kind="ecdf", x="total_bill", col="time", hue="smoker")
```

```python
tips = sns.load_dataset("tips")
sns.catplot(data=tips,kind="swarm", x="day", y="total_bill", hue="smoker")
```
